#renaming columns

```python
df.rename(columns = {'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State','region_txt':'Region','attacktype1_txt':'Attack_type','target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
New_df = df[['Year','Month','Day','Country','State','Region','city','latitude','longitude','Attack_type','Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
```

#handling missing data 

```python
New_df.describe()
```

```python
New_df.isnull().sum()
```

#replacing blank values of each column in new data set with zero

```python
New_df['State'].fillna(0,inplace=True)
New_df['city'].fillna(0,inplace=True)
New_df['latitude'].fillna(0,inplace=True)
New_df['longitude'].fillna(0,inplace=True)
New_df['Killed'].fillna(0,inplace=True)
New_df['Wounded'].fillna(0,inplace=True)
New_df['Target'].fillna(0,inplace=True)
New_df['Summary'].fillna(0,inplace=True)
New_df['Motive'].fillna(0,inplace=True) 
New_df['Group'].fillna(0,inplace=True)
New_df['Weapon_type'].fillna(0,inplace=True)     
```

#Country Wise Terrorist Attacks

```python
fig,axes = plt.subplots(figsize=(10,8),nrows=1,ncols=2)
sns.barplot(x = df['Country'].value_counts()[:20].values, y = df['Country'].value_counts()[:20].index,ax=axes[0],palette = 'coolwarm_r');
axes[0].set_title('Country Wise Terrorist Attacks')
sns.barplot(x=df['Region'].value_counts().values,y=df['Region'].value_counts().index,ax=axes[1])
axes[1].set_title('Region Wise Terrorist Attacks')
fig.tight_layout()
plt.show()
```

#Year Wise Terrorist Activities

```python
pd.crosstab(df.Year, df.Region).plot(kind='area',figsize=(10,5))
plt.title('Year wise Terrorist Activities by Region')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
plt.bar(x='Year',height='Killed',data=New_df,color='orange')
plt.xticks(rotation=90)  
plt.title('Year Wise Terrorist Activities')
```

```python
Year=New_df.Year.value_counts().to_dict()
rate=((Year[2017]-Year[1970])/Year[2017])*100
print('A total of', Year[1970],'attacks occured in 1970 and a total of',Year[2017],'attacks occured in the year 2017.')
print('The total number of attacks from 1970 to 2017 has increased by',np.round(rate,0),'percent.')
```

#methods of attack

```python
sns.displot(New_df['Attack_type'],color='red')
plt.xticks(rotation=90)
plt.xlabel('Method')
plt.title('Methods of Attack')
plt.figure(figsize=(10,5))
```

#most active terrorist organizations

```python
plt.subplots(figsize=(8,6))
sns.barplot(y=df['Group'].value_counts()[1:12].index,x=df['Group'].value_counts()[1:12].values,palette='deep')
plt.title('Most Active Terrorist Organizations')
plt.show()
```

Insights found:

Bombing/Explosion is the most frequent method of attack by terrorists.
